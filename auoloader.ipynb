```python
%sql
create schema netflix_catalog.net_schema
```

```python
cl = "abfss://silver@netflixadlsraj.dfs.core.windows.net/"
#checkpoint Location
```

```python
df = spark.readStream\
    .format("cloudFiles")\
    .option("cloudFiles.format", "csv")\
    .option("cloudFiles.schemaLocation", cl)\
    .load("abfss://raw@netflixadlsraj.dfs.core.windows.net/")
```

```python
df.display()
```

```python
df.writeStream\
  .option("checkpointLocation", cl)\
  .trigger(processingTime='10 seconds')\
  .start("abfss://bronze@netflixadlsraj.dfs.core.windows.net/netflix_titles")    
```
